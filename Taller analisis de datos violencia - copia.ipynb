```python
# Importar base de datos 
import pandas as pd

# Cambia el nombre del archivo a algo más simple si puedes
ruta = r'C:\Users\andre\OneDrive\Escritorio\Analisis_Cari\violencia.xlsx'

# Leer el archivo Excel
df = pd.read_excel(ruta)

# Mostrar las primeras filas
print(df.head())
```

```python
#Pasar variables a minusculas para evitar errores (Ejemplo variable sexo de agresor donde exiten datos como HOMBRES y Hombres)
df = df.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)
```

```python
# Estandarizar nombre del mes
df['Mes del hecho'] = df['Mes del hecho'].str.lower().str.strip()

# Diccionario de meses
meses = {
    'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4,
    'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8,
    'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12
}

# Crear columna de periodo (AÑO-MES) sin día porque este dato no lo tengo
df['periodo'] = pd.to_datetime(
    df['Año del hecho'].astype(str) + '-' + df['Mes del hecho'].map(meses).astype(str),
    format='%Y-%m'
).dt.to_period('M')
```

```python
# Reviso como quedo mi variable periodo (si estan bien el codigo)
print(df[['Año del hecho', 'Mes del hecho', 'periodo']].head())
```

```python
# Elimino columnas que no me aportan mayor información
columnas_a_eliminar = [
    'Grupo Mayor Menor de Edad',
    'Edad judicial',
    'Rango de Hora del Hecho X 3 Horas',
    'Localidad del Hecho',
    'Contexto de violencia',
    'Condición de la Víctima',
    'Medio de Desplazamiento o Transporte',
    'Servicio del Vehículo',
    'Clase o Tipo de Accidente',
    'Objeto de Colisión',
    'Servicio del Objeto de Colisión',
    'Días de Incapacidad Medicolegal'
]

df.drop(columns=columnas_a_eliminar, inplace=True)
```

## Análisis Exploratorio

```python
departamentos_caribe = [
    'Atlántico', 'Bolívar', 'Cesar', 'Córdoba',
    'La Guajira', 'Magdalena', 'Sucre', 'San Andrés'
]
```

```python
# Estandarizar nombre de departamento
df['Departamento del hecho DANE'] = df['Departamento del hecho DANE'].str.strip().str.title()

# Filtrar solo los departamentos caribeños
df_caribe = df[df['Departamento del hecho DANE'].isin(departamentos_caribe)]
```

```python
agrupado = df_caribe.groupby(['Año del hecho', 'Departamento del hecho DANE']).size().reset_index(name='Casos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener lista ordenada de años
anios = sorted(df_caribe['Año del hecho'].dropna().unique())

# Crear un gráfico por año
for anio in anios:
    datos_anio = agrupado[agrupado['Año del hecho'] == anio]

    plt.figure(figsize=(10, 5))
    sns.barplot(data=datos_anio, x='Departamento del hecho DANE', y='Casos', palette='Blues_d')
    plt.title(f'Casos de violencia en la región Caribe - Año {anio}')
    plt.xlabel('Departamento')
    plt.ylabel('Número de casos')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Limpieza de texto
df['Sexo del Presunto Agresor'] = df['Sexo del Presunto Agresor'].str.strip().str.lower()
df['Sexo de la victima'] = df['Sexo de la victima'].str.strip().str.lower()
```

```python
tabla_sexos = pd.crosstab(df['Sexo de la victima'], df['Sexo del Presunto Agresor'])
print(tabla_sexos)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(tabla_sexos, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Relación entre sexo del agresor y sexo de la víctima')
plt.xlabel('Sexo del presunto agresor')
plt.ylabel('Sexo de la víctima')
plt.tight_layout()
plt.show()
```

```python
# Elegir el año de análisis
anio = 2022

# Filtrar por ese año
df_anio = df[df['Año del hecho'] == anio].copy()

# Limpiar texto
df_anio['Sexo del Presunto Agresor'] = df_anio['Sexo del Presunto Agresor'].str.strip().str.lower()
df_anio['Sexo de la victima'] = df_anio['Sexo de la victima'].str.strip().str.lower()

# Tabla cruzada
tabla = pd.crosstab(df_anio['Sexo de la victima'], df_anio['Sexo del Presunto Agresor'])

# Visualización
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(tabla, annot=True, fmt='d', cmap='coolwarm')
plt.title(f'Relación sexo víctima vs agresor - Año {anio}')
plt.xlabel('Sexo del presunto agresor')
plt.ylabel('Sexo de la víctima')
plt.tight_layout()
plt.show()
```

```python
# Limpiar columnas de texto
df['Sexo del Presunto Agresor'] = df['Sexo del Presunto Agresor'].str.strip().str.lower()
df['Sexo de la victima'] = df['Sexo de la victima'].str.strip().str.lower()

# Agrupar por año, sexo de víctima y sexo de agresor
agrupado = df.groupby(['Año del hecho', 'Sexo de la victima', 'Sexo del Presunto Agresor']).size().reset_index(name='Casos')

# Filtrar solo combinaciones más comunes (opcional)
comunes = agrupado[
    (agrupado['Sexo de la victima'].isin(['hombre', 'mujer'])) &
    (agrupado['Sexo del Presunto Agresor'].isin(['hombre', 'mujer']))
]

# Crear columna combinada para el gráfico
comunes['Relacion'] = comunes['Sexo de la victima'] + ' - ' + comunes['Sexo del Presunto Agresor']

# Visualizar
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.lineplot(data=comunes, x='Año del hecho', y='Casos', hue='Relacion', marker='o')
plt.title('Evolución anual de casos según sexo de víctima y agresor')
plt.xlabel('Año')
plt.ylabel('Número de casos')
plt.xticks(comunes['Año del hecho'].unique())  # mostrar todos los años
plt.legend(title='Víctima - Agresor')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar y limpiar
tabla_circ_sexo_agresor = pd.crosstab(
    df['Circunstancia del Hecho'].str.strip().str.lower(),
    df['Sexo del Presunto Agresor'].str.strip().str.lower()
)

# Opción: mostrar solo las 10 circunstancias más frecuentes
top_circ = tabla_circ_sexo_agresor.sum(axis=1).sort_values(ascending=False).head(10).index
tabla_top = tabla_circ_sexo_agresor.loc[top_circ]

# Graficar
plt.figure(figsize=(10, 6))
sns.heatmap(tabla_top, annot=True, fmt='d', cmap='YlOrBr')
plt.title('Circunstancia del hecho vs Sexo del presunto agresor')
plt.xlabel('Sexo del presunto agresor')
plt.ylabel('Circunstancia del hecho')
plt.tight_layout()
plt.show()
```
